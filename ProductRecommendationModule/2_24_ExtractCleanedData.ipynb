```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('giftcart_cleaned_final_proper.csv')

# Select only Handle, Title, and Description columns
df_filtered = df[['Handle', 'Title', 'Description']].copy()

# Remove rows where all three columns are empty/null
df_filtered = df_filtered.dropna(how='all')

# Also remove rows where all three are empty strings
df_filtered = df_filtered[~((df_filtered['Handle'].fillna('').str.strip() == '') & 
                             (df_filtered['Title'].fillna('').str.strip() == '') & 
                             (df_filtered['Description'].fillna('').str.strip() == ''))]

print(f"Total rows extracted: {len(df_filtered)}")
df_filtered.head(10)
```

```python
# Add auto-incrementing ID column
df_filtered.insert(0, 'id', range(1, len(df_filtered) + 1))

# Save to CSV file
df_filtered.to_csv('giftcart_cleaned_products_6000.csv', index=False)

print(f"Saved {len(df_filtered)} rows to giftcart_cleaned_products_6000.csv")
df_filtered.head(10)
```
