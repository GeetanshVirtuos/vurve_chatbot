# Filter Data by Label

This notebook loads the labeled data CSV, filters rows with label 2 (gift suggestions), and saves them to a new file.

```python
import pandas as pd
```

```python
# Load the CSV file
df = pd.read_csv('../IntentClassificationModule/more_labeled_data.csv', header=None, names=['text', 'label'])

# Display basic info
print(f"Total rows: {len(df)}")
print(f"\nLabel distribution:")
print(df['label'].value_counts().sort_index())
print(f"\nFirst few rows:")
df.head()
```

```python
# Filter rows with label 2
filtered_df = df[df['label'] == 2]

print(f"Rows with label 2: {len(filtered_df)}")
print(f"\nFirst few filtered rows:")
filtered_df.head(10)
```

```python
# Save to a new CSV file with quotes preserved
import csv
output_path = 'label_2_filtered.csv'
filtered_df.to_csv(output_path, index=False, header=False, quoting=csv.QUOTE_NONNUMERIC)

print(f"✓ Filtered data saved to: {output_path}")
print(f"✓ Total {len(filtered_df)} rows saved (with text in quotes)")
```
