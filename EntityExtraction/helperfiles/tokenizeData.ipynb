# Filter Data by Label

This notebook loads the labeled data CSV, filters rows with label 2 (gift suggestions), and saves them to a new file.

```python
import pandas as pd
from gensim.utils import simple_preprocess
```

```python
# Load the CSV file
df = pd.read_csv('labeled_data.csv', header=0)

# Display basic info
print(f"Total rows: {len(df)}")
print(f"\nLabel distribution:")
print(df['label'].value_counts().sort_index())
print(f"\nFirst few rows:")
df.head()
```

```python
# Apply gensim simple_preprocess to create tokens column
df['tokens'] = df['sentence'].apply(lambda x: '|'.join(simple_preprocess(x)))

print("Sample tokenized sentences:")
print(df[['sentence', 'tokens']].head())
```

```python
# Save to CSV with headers
import csv
output_path = 'label_token_data.csv'

# Reorder columns to have sentence, label, tokens
df_output = df[['sentence', 'label', 'tokens']]
df_output.to_csv(output_path, index=False, header=True, quoting=csv.QUOTE_NONNUMERIC)

print(f"✓ Filtered data saved to: {output_path}")
print(f"✓ Total {len(df)} rows saved")
print(f"✓ Columns: sentence, label, tokens")
```
