```python
import pandas as pd

df = pd.read_csv('TshirtLaptopData.csv', sep='|', encoding='cp1252')
df.head()
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)
df_queries = df[['text']]
```

```python
# Randomly sample 3000 rows
df_sampled = df[['text']].sample(n=3000, random_state=42)

# Add the label column with value 2
df_sampled['label'] = 2

# Save to CSV with quotes only around text, not the number
import csv
df_sampled.to_csv('sampled_queries.csv', index=False, header=False, quoting=csv.QUOTE_NONNUMERIC)

print(f"Saved {len(df_sampled)} rows to sampled_queries.csv")
df_sampled.head()
```
